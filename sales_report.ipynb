```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_excel('sales_data.xlsx', engine='openpyxl')
print(data)


# Data processing
# Add your data processing code here

# Data processing
# Calculate total sales
data['Total Sales'] = data['Quantity Sold'] * data['Unit Price']

# Extract month and year from the date
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year

# Group by month and year to get monthly sales
monthly_sales = data.groupby(['Year', 'Month'])['Total Sales'].sum().reset_index()

# Get top-selling products
top_products = data.groupby('Product')['Total Sales'].sum().nlargest(10).reset_index()

# Data visualization
# Bar chart for sales by product
plt.figure(figsize=(10, 6))
sales_by_product = data.groupby('Product')['Total Sales'].sum().reset_index()
plt.bar(sales_by_product['Product'], sales_by_product['Total Sales'])
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

# Histogram for unit price distribution
plt.figure(figsize=(10, 6))
plt.hist(data['Unit Price'], bins=10, color='skyblue', edgecolor='black')
plt.title('Unit Price Distribution')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.show()

# Create a report with insights
# Add your insights and interpretations here
```
